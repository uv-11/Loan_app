```python
import pandas as pd
import numpy as np
#!pip install openpyxl# python requores this openpyxl library to open xlsx files 
```

```python
data = pd.read_csv(r"Loan_default_1500.csv")

data = data[['Credit_Score', 'income', 'loan_amount', 'dtir1', 'Status']]
data.head()
```

```python
data.isna().sum()
data.info()
```

```python
data['Credit_Score'].fillna(data['Credit_Score'].mean(), inplace=True)
data['income'].fillna(data['income'].median(), inplace=True)
data['loan_amount'].fillna(data['loan_amount'].mean(), inplace=True)
data['dtir1'].fillna(data['dtir1'].mean(), inplace=True)
```

```python

data.info()
data.isna().sum()
```

```python
data['income'] = data['income'].fillna(data['income'].median())
data['dtir1'] = data['dtir1'].fillna(data['dtir1'].median())
```

```python
data.isna().sum()
```
